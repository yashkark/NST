```python
import json
```

```python
with open('optimizers_losses.json', 'r') as file:
    data = json.load(file)
```

```python
Adam_e5000 = data["Adam"]["e5000"]
LBFGS_e5000 = data["LBFGS"]["e5000"]
LBFGSv2_e5000 = data["LBFGSv2"]["e5000"]
```

```python
Adam_e5000[0], LBFGS_e5000[0], LBFGSv2_e5000[0]
```

```python
len(Adam_e5000), len(LBFGS_e5000), len(LBFGSv2_e5000)
```

```python
Adam_e5000[-1], LBFGS_e5000[-1], LBFGSv2_e5000[-1]
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_values = list(range(1, 5001))

# Create line plots for both lists
plt.plot(x_values, np.log(Adam_e5000), label='Adam')
plt.plot(x_values, np.log(LBFGS_e5000), label='LBFGS')
plt.plot(x_values, np.log(LBFGSv2_e5000), label='LBFGSv2')

# Adding title, labels and legend
plt.title('Line Plot for Loss value from Adam and LBFGS Optimizers')
plt.xlabel('Epochs')
plt.ylabel('log(Loss)')
plt.legend()

# Show the plot
plt.show()
```
