# Experiment 5

In this experiment we attempt to combine to artistic styles. We use one content image and two style using images. We then modify the loss function to add a third parameter gamma. We focus on changing beta and gamma to understand how they influence combination of styles

The updated loss function is

$\mathbf{L}=\alpha (Content\:Loss) + \beta( Style\:Loss_1) + \gamma( Style\:Loss_2)$

where the Content Loss and Style Loss are defined as

$Content\:Loss = \frac{(target\:features \:- \:content\:features)^2}{2}$

where target features are the features of the new image generated at every iteration of gradient descent.



## Experiment 5.1

```python
alpha=0.2
beta=0.8
gamma=0.8
```

```python
obj = StyleTransfer(content_image3, style_image2,style_image3)
```

```python
exp_name = "Experiment_5.1"
obj.set_exp(exp_name)
obj.show_images()
```

```python
obj.run(show_every=1000, steps = 5000, alpha=alpha, beta=beta,gamma=gamma)
```

```python
create_gif(exp_name)
```
