# Experiment 1

In this experiment we are manipulating two coefficients alpha and beta which effect the amount of content loss and style loss respectively.

Essentially the loss function is

$\mathbf{L}=\alpha (Content\:Loss) + \beta( Style\:Loss)$

where the Content Loss and Style Loss are defined as

$Content\:Loss = \frac{(target\:features \:- \:content\:features)^2}{2}$

where target features are the features of the new image generated at every iteration of gradient descent.



## Experiment 1.1

```python
alpha=1
beta=1
```

### Experiment 1.1.1

```python
obj = StyleTransfer(content_image1, style_image4)
```

```python
exp_name = "Experiment_1.1.1"
obj.set_exp(exp_name)
obj.show_images()
```

```python
obj.run(alpha=alpha, beta=beta, opt='LBFGS')
loss=obj.get_loss()
loss = [t.item() for t in loss]
#losses[exp_name]=loss
```

```python
create_gif(exp_name)
```

```python
display(Image(filename="Outputs/{0}/gif.gif".format(exp_name)))
```
